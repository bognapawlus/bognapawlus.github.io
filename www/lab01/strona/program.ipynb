```python
##1. downloading the website
from bs4 import BeautifulSoup
import requests
from bs4 import BeautifulSoup
response = requests.get('https://www.alcazar.in/blog/a-complete-dog-breed-list')
print(response.text)
html_content = response.text
```

```python
##2. converting to BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')
print(soup.prettify())
```

```python
list_of_dogs = soup.find_all('td')
list_of_dogs
```

```python
n = len(list_of_dogs)
n
```

```python
#list with
column_names = list_of_dogs[0:7]
column_names
```

```python
i = 7
with open('strona3.md', 'w') as f:
    while i < n:
        start = i
        end = i + 7

        for el in range(start, end):
            f.write(f"{list_of_dogs[el].text}\t\t")
        f.write("\n")

        i = end
```
